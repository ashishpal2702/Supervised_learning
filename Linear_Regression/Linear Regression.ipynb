## Data Preprocessing

```python
housing_data.isna().sum()
```

```python
housing_data.corr()
```

```python
housing_data_num = housing_data
for col in housing_data_num.columns:
    #print(housing_data_num[col].dtypes)
    if housing_data_num[col].dtypes == 'object':
        housing_data_num.drop(col, inplace = True , axis = 1)
housing_data_num.info()   
```

```python
### Train Test Split: 

X = housing_data_num.drop('SalePrice', axis =1 )
Y = housing_data_num['SalePrice']

from sklearn.model_selection import train_test_split
X_train , X_test , Y_train, Y_test = train_test_split(X,Y, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
X_train_transformed = sc.fit_transform(X_train)
```

```python
X_train_transformed.shape
```

```python
X_train_transformed
```

```python
X_test_transformed = sc.transform(X_test)
```

```python
X_test_transformed
```

## Model - Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
## Model Fitting / Training 

lr.fit(X_train_transformed, Y_train)
```

```python
print("Training Score ", lr.score(X_train_transformed, Y_train))
```

```python
print("Testing Score ", lr.score(X_test_transformed, Y_test))
```

```python
prediction = lr.predict(X_test)
```
